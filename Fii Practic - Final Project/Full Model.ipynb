### 1. Age
We are trying to replace the age missing values based on the correlations with other features. Let's check and see the highest correlations with age feature.

```python
correlations = train.corr().abs().unstack().sort_values(ascending = False)
correlations.Age
```

We notice that *Age* and *Pclass* have a high correlation. It would be an idea to fill the Age missing values based on passenger classes. An idea would be to replace the missing values in Age with median. Though, using the age median of the whole dataset would not be a good idea, so we could make use of the correlation shown above and fill the **Age** column with the median of Pclass groups. 

Also, in order to be more accurate and precise, *Sex* feature will be used as the second level of groupby while filling the missing Age values. As seen from below, Pclass and Sex groups have distinct median Age values. When passenger class increases, the median age for both males and females also increases. However, females tend to have slightly lower median Age than males. The median ages below are used for filling the missing values in Age feature.

```python
age_by_pclass_sex = train.groupby(['Sex', 'Pclass']).median()['Age']

for p_class in range(1, 4):
    for sex in ['female', 'male']:
        print('Median age of Pclass {} {}s: {}'.format(p_class, sex, age_by_pclass_sex[sex][p_class]))

print()
print('Median age of all passengers: {}'.format(train['Age'].median()))
```

Now, we will fill the age missing values.

```python
train['Age'] = train.groupby(['Sex', 'Pclass'])['Age'].apply(lambda x: x.fillna(x.median()))
```

```python
def get_pclass_dist(df):
    
    # Creating a dictionary for every passenger class count in every deck
    deck_counts = {'A': {}, 'B': {}, 'C': {}, 'D': {}, 'E': {}, 'F': {}, 'G': {}, 'M': {}, 'T': {}}
    decks = df.columns.levels[0]    
    
    for deck in decks:
        for pclass in range(1, 4):
            try:
                deck_counts[deck][pclass] = df[deck][pclass][0]
            except KeyError:
                deck_counts[deck][pclass] = 0
                
    df_decks = pd.DataFrame(deck_counts)    
    deck_percentages = {}

    # Creating a dictionary for every passenger class percentage in every deck
    for col in df_decks.columns:
        deck_percentages[col] = [(count / df_decks[col].sum()) * 100 for count in df_decks[col]]
        
    return deck_counts, deck_percentages

def display_pclass_dist(percentages):
    
    df_percentages = pd.DataFrame(percentages).transpose()
    deck_names = ('A', 'B', 'C', 'D', 'E', 'F', 'G', 'M', 'T')
    bar_count = np.arange(len(deck_names))  
    bar_width = 0.65
    
    pclass1 = df_percentages[0]
    pclass2 = df_percentages[1]
    pclass3 = df_percentages[2]
    
    plt.figure(figsize=(20, 10))
    plt.bar(bar_count, pclass1, color='darkmagenta', edgecolor='white', width=bar_width, label='Passenger Class 1')
    plt.bar(bar_count, pclass2, bottom=pclass1, color='hotpink', edgecolor='white', width=bar_width, label='Passenger Class 2')
    plt.bar(bar_count, pclass3, bottom=pclass1 + pclass2, color='lightsalmon', edgecolor='white', width=bar_width, label='Passenger Class 3')

    plt.xlabel('Deck', size=15, labelpad=20)
    plt.ylabel('Passenger Class Percentage', size=15, labelpad=20)
    plt.xticks(bar_count, deck_names)    
    plt.tick_params(axis='x', labelsize=15)
    plt.tick_params(axis='y', labelsize=15)
    
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1), prop={'size': 15})
    plt.title('Passenger Class Distribution in Decks', size=18, y=1.05)   
    
    plt.show()    
```

```python
all_deck_count, all_deck_per = get_pclass_dist(df_all_decks)
display_pclass_dist(all_deck_per)
```

### Statistical Tests

#####  1) T-Test (Measuring male - female ratio)

The T-test (aka Student’s T Test) checks whether a sample mean differs from the population mean. Since Sex have a high correlation with the dependent variable Survived, we can test to see if the mean for Sex that survived differ from the mean Sex that did not survive.

***Hypothesis Testing***: Is there a significant difference in the mean for Sex between the passenger who survived and passenger who did not survive?

**(H0)**: There is no difference in the mean for Sex between the passenger who survived and passenger who did not survive. 

**(H1)**: There is a difference in the mean for Sex between the passenger who survived and those who did not.

```python
mean_survived_male = train[train['Sex'] == 'male']['Survived'].mean()
mean_survived_female = train[train['Sex'] == 'female']['Survived'].mean()

print("The average survival rate for male passengers is {}".format(mean_survived_male))
print("The average survival rate for female passengers is {}".format(mean_survived_female))
```

So we can see that female and 1st class passengers had better chances for survival. But is it a statistically significant result?
We can use hypothesis testing techniques to find this out.

```python
male = train[train['Sex'] == 'male']['Survived']

female = train[train['Sex'] == 'female']['Survived']
```

T-test can be calculated by using the following formula <img src="https://miro.medium.com/max/932/1*1ZUnA4eR5J2WEGhDVPDkEw.png" width = 500px/> where 
$M_x$ and $M_y$ are the mean values of the two samples of male and female, 

$n_x$ and $n_y$ are the sample space of the two samples, 

$S$ is the standard deviation. 


```python
male_mean = male.mean()
female_mean = female.mean()

# get sample variance
male_var = statistics.variance(male)
female_var = statistics.variance(female)

# get t-value
t = (female_mean - male_mean)/math.sqrt(female_var/len(female) + male_var/len(male))
print('T-value is {t}'.format(t = t))

# degrees of freedom
df = len(male) + len(female) - 2

# calculate the critical t-value
alpha = 0.05
cv = stats.t.ppf(1.0 - alpha, df)
print('Critical t-value is {}'.format(cv))

# find p-value
p = 1 - stats.t.cdf(t,df=df)
print('P-value is {p}'.format(p = p))
```

We'll reject the null hypothesis because:
- the t-value is great than the critical t-value
- the P-value is less than 5%


The test concludes that there is a statistically significant difference between the two populations. So female passengers had statistically significant more chances to survive.


#####  2) ANOVA (Measuring pclasses ratio)

The analysis of variance (ANOVA) can be thought of as an extension to the t-test. The independent t-test is used to compare the means of a condition between 2 groups. ANOVA is used when one wants to compare the means of a condition between 2+ groups. We also noticed from our correlation matrix that Pclass and Survived have a high correlation. Let's check if there's a significant difference in the mean for Pclass1, Pclass2 and Pclass3.

***Hypothesis Testing***: Is there a significant difference in the mean for Pclass between the passenger who survived and passenger who did not survive?

**(H0)**: There is no difference in the mean for Pclass between the passenger who survived and passenger who did not survive. 

**(H1)**: There is a difference in the mean for Pclass between the passenger who survived and those who did not.

```python
mean_survived_1stclass = train[train['Pclass'] == 1]['Survived'].mean()
mean_survived_2ndclass = train[train['Pclass'] == 2]['Survived'].mean()
mean_survived_3rdclass = train[train['Pclass'] == 3]['Survived'].mean()

print("The average survival rate for 1st class passengers is {}".format(mean_survived_1stclass))
print("The average survival rate for 2nd class passengers is {}".format(mean_survived_2ndclass))
print("The average survival rate for 3rd class passengers is {}".format(mean_survived_3rdclass))
```

```python
anova_stats = stats.f_oneway(train[train['Pclass'] == 1]['Survived'], 
             train[train['Pclass'] == 2]['Survived'],
             train[train['Pclass'] == 3]['Survived'])
```

```python
if anova_stats[1] < 0.05:
    print("We are rejecting the null hypothesis")
else:
    print("We are accepting the null hypothesis")
```

The test concludes that there is a statistically significant difference between the three populations. So 1st class passengers had statistically significant more chances to survive.

We can now compare our models and to choose the best one for our problem.

```python
models = pd.DataFrame({
    'Model': ['Logistic Regression', 'Support Vector Machines', 'Stochastic Gradient Descent', 
              'Decision Tree', 'Random Forest (GridSearchCV)', 
              'XGB (GridSearchCV)', 'Gradient Boosting',  'AdaBoost', 'kNN (GridSearchCV)', 'Naive Bayes', 'Perceptron', 'MultiLayer Perceptron']})
```

```python
classifiers = {
    'Logistic Regression': logisticRegression,
    'Support Vector Machines': svm,
    'Stochastic Gradient Descent': sgd,
    'Decision Tree': decisionTree,
    'Random Forest (GridSearchCV)': model,
    'XGB (GridSearchCV)': xgb,
    'Gradient Boosting': gradient_boosting, 
    'AdaBoost': ada, 
    'kNN (GridSearchCV)': knn1,
    'Naive Bayes': gaussian,
    'Perceptron': perc,
    'MultiLayer Perceptron': mp,
}

cross_val_on_train = []
cross_val_on_test = []

for clf in classifiers:
    scores_train = cross_val_score(classifiers[clf], X_train, y_train, cv = 10, scoring = 'accuracy').mean()
    cross_val_on_train.append(scores_train)
    scores_test = cross_val_score(classifiers[clf], X_test, y_test, cv = 10, scoring = 'accuracy').mean()
    cross_val_on_test.append(scores_test)
    
print(cross_val_on_train)
print(cross_val_on_test)
```
